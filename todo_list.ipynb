```python
from datetime import datetime
import json
class Database(object):
    def __init__(self) -> None:
        self.tem_database = {}

    def load_from_database_file(self,file_path):
        with open(file_path, 'r') as file:
            data = json.load(file)
            self.data = data
            return self.data
        
    def get_database(self):
        self.database = self.data
    
    def write_to_database_file(self,file_path):
        with open(file_path, 'w') as file:
            json.dump(self.database,file)

    
    def write_to_recyclebin(self):
        file_path = "recyclebin.txt"
        with open(file_path, 'w') as file:
            json.dump(self.recyclebin,file)


class Task(object):
    def __init__(self,task_name,task_priority,task_note,task_status = False) -> None:
        self.task_name = task_name
        self.task_priority = task_priority
        self.task_note = task_note
        self.task_status = task_status
    

    def __str__(self) -> str:
        status = "Done" if self.task_status else "Not Done"
        return f"{self.task_name}, {self.task_priority}, {self.task_note}, {status}"
class AlertMeassages(object):
    current_time = datetime.now().strftime("%d-%m-%Y  %H:%M:%S")
    def __init__(self) -> None:
        pass
    def alert_log(self,alert_log_meassage):
        alert_meassage = f"{self.current_time} \n{alert_log_meassage}"
        with open('alert_log.txt','a') as file:
            file.write('\n' + alert_meassage + '\n')
 
    def list_alert_task(self):
        file = open('alert_log.txt','a')
        alert_meassage = f"{self.current_time}: Below is the list of your assigned task thank you. \n "
        file.write('\n' + alert_meassage +'\n')
        for task_key, task_value in self.database.items():
            print(task_key,task_value,file=file)

class ToDOList(AlertMeassages,Database):
    def add_task(self,task):
        file_path = "database_file.txt"
        Database.load_from_database_file(self,file_path)
        Database.get_database(self)
        property_data = {}
        new_task = task
        convert = str(new_task)
        list = convert.split(",")
        task_name = list[0]
        task_name = task_name.upper()
        task_priority = list[1]
        task_note = list[2]
        task_status = list[3]
        property_data["TASK PRIORITY"] = task_priority
        property_data["TASK NOTE"] = task_note
        property_data["TASK STATUS"] = task_status
        self.database[task_name] = property_data
        Database.write_to_database_file(self,file_path)
  
    
    def remove_task(self,task_name):
        file_path1 = "database_file.txt"
        file_path2 = "recyclebin.txt"
        Database.load_from_database_file(self,file_path1)
        Database.get_database(self)
        recycle = self.database.pop(task_name)
        Database.write_to_database_file(self,file_path1)
        Database.load_from_database_file(self,file_path2)
        Database.get_database(self)
        self.database[task_name] = recycle
        Database.write_to_database_file(self,file_path2)
        alert_log_meassage = f"Your task '{task_name}' has been successfully deleted from your tasks list and stored on the recyclebin"
        AlertMeassages.alert_log(self,alert_log_meassage)

    def completed_task(self,task_name):
        file_path1 = "database_file.txt"
        Database.load_from_database_file(self,file_path1)
        Database.get_database(self)
        var = self.database[task_name]
        var2 = var["TASK STATUS"] = "Done"
        Database.write_to_database_file(self,file_path1)
        
    def display_tasks(self):
        file_path = "database_file.txt"
        Database.load_from_database_file(self,file_path)
        Database.get_database(self)
        if self.database == {}:
            alert_log_meassage = "Your tasks list is empty thank you."
            AlertMeassages.alert_log(self,alert_log_meassage)
        else:
            AlertMeassages.list_alert_task(self)
```
